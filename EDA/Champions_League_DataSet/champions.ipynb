```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AllTimeRankingByClub.csv", encoding="utf-16")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
top_20 = df.groupby("Club")["Titles"].sum().sort_values(ascending= False)[:20]
top_20
```

```python
top_20.index
```

```python
top_20.values
```

```python
plt.figure(figsize=(10,8))
plt.bar(top_20.index, top_20.values)
plt.xticks(rotation= 90);
```

```python
plt.figure(figsize=(10,8))
plt.barh(top_20.index, top_20.values);
```

```python
df.Country.value_counts().index
```

```python
bool_winners = df.groupby("Country")["Titles"].sum().sort_values(ascending=False) >0
bool_winners
```

```python
winners =df.groupby("Country")["Titles"].sum().sort_values(ascending=False)[bool_winners]
winners
```

```python
x = winners.index
x
```

```python
y = winners.values
y
```

```python
# explode_list = [0.1, 0.1, 0.2, 0.3]
plt.figure(figsize=(8,4))
mylabels = x

plt.pie(y, labels = mylabels, labeldistance = 0.7, autopct="%.1f", startangle = 0, pctdistance=0.6
        )
plt.legend(loc=(0))
plt.show()
```
