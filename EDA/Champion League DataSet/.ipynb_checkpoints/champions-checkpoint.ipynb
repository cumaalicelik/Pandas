```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AllTimeRankingByClub.csv", encoding="utf-16")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe().T
```

```python
# df.groupby("Club")["Titles"].sum().sort_values(ascending=False)[:20]
plt.figure(figsize=(10,6))

(df.groupby("Club")["Titles"].sum().sort_values(ascending=False)[:20]).plot(kind='bar')

plt.xticks(rotation=60, ha="right")
plt.xlabel("Titles");
```

```python
#horizontal

plt.figure(figsize=(10,6))
(df.groupby("Club")["Titles"].sum().sort_values(ascending=False)[:20]).plot(kind='barh')
plt.xlabel("Titles");
```

```python
df.groupby("Club")["Titles"].sum().sort_values(ascending=False)
```

```python
has_title = df[df.Titles >=1]
has_title
```

```python
has_title = has_title.sort_values('Titles', ascending=False) # we now can sort the below chart in order!
has_title
```

```python
plt.figure(figsize=(12,7))

sns.barplot(data=has_title, x="Club", y="Titles")
plt.xticks(rotation = 75, ha = 'right',   
           fontsize = 12, color = 'blue') # ha = 'right' will solve overlapping problem

plt.yticks(range(15)) # we give a range to see all the ytick numbers!

plt.show()
```

```python
# we will find the most participated and zero title owners 

bad_luck_club = df[(df.Participated >=25 )& (df.Titles ==0)].sort_values('Participated', ascending=False)
bad_luck_club
```

```python
fig = plt.figure(figsize=(12,6))

sns.barplot(data=bad_luck_club, x="Club", y="Participated")
plt.xticks(rotation = 75, ha = 'right',   
           fontsize = 12, color = 'blue') # ha = 'right' will solve overlapping problem

plt.rcParams['axes.facecolor'] = 'black' # changing the background color of axes

# plt.yticks(range(39))

fig.set_facecolor('yellow') # it changes the background of chart



plt.show()
```

```python
#Is there any team that has 10 participation and title?

df[(df.Participated <=10) & (df.Titles >= 1)]
```

```python
sns.scatterplot(data=has_title, y="Participated", x="Titles", hue="Club")

import matplotlib as mpl
mpl.rcParams['text.color'] = 'white'

plt.legend(loc=(1.1,0))
plt.show()
```

```python
# first 5 countries piechart

plt.pie(x=has_title.Titles[:5] ,labels=has_title.Country[:5], explode=(.1,0,0,0,0), startangle=0, autopct='%1.1f%%')

plt.legend(loc=(.9,.7))

plt.show()
```

***What is the percentence of the Turkish clubs?***

```python
tur_par = df[df.Country =="TUR"].Participated.sum()
tur_par
```

```python
tur_per = tur_par / df.Participated.sum()
```

```python
print(f"The percentage of the Turkish culubs participating in UEFA CL: {round(tur_per *100, 2)}%")
```

```python
country = pd.read_csv("AllTimeRankingByCountry.csv")
country.head()
```

```python
#drop unnecessary column

country.drop(['Unnamed: 0'], axis=1, inplace=True)
country.head()
```

```python
# null values?

country.isna().sum()[country.isna().sum() > 0]
```

```python
country.describe().T
```

```python
# titled countries

country_has_title = country[country.Titles > 0].sort_values("Titles", ascending=False)
country_has_title
```

```python
fig = plt.figure(figsize=(12,6))

sns.barplot(data=country_has_title, x="Country", y="Titles")
plt.xticks(rotation = 45, ha = 'right',   
           fontsize = 15, color = 'blue') # ha = 'right' will solve overlapping problem

plt.rcParams['axes.facecolor'] = 'lightgray' # changing the background color of axes

plt.yticks(range(20))

fig.set_facecolor('yellowgreen') # it changes the background of chart

plt.grid(axis="y", color="purple") # we are just drawing grids on y axis, default is for both

mpl.rcParams['text.color'] = 'red'

plt.title("Numbers of the Titles Countries Have", color="purple",fontsize=20)

plt.show()
```

```python
goal_detail = pd.read_csv("PlayerGoalDetails.csv")
goal_detail
```

```python
# we will now examine player appearances

app = pd.read_csv("PlayerAppearDetails.csv")
app.head()
```

```python
app.drop(["Unnamed: 0"], axis= 1, inplace=True)
app.head()
```

```python
app.groupby("Player")["Appearances"].sum()

# we can see that the appearances column is not int!!!
```

```python
app.Appearances.astype(int)

# we see that there is (10), so we firstly get rid of it
```

```python
app['Appearances'] = app['Appearances'].str.replace(r"\(|\)", "").astype(int)
```

```python
app.Appearances.dtypes
```

```python
app.groupby("Player")["Appearances"].sum().sort_values(ascending=False)[:20]
```

```python
# top 25 app for players in CL

plt.figure(figsize=(10,6))
app.groupby("Player")["Appearances"].sum().sort_values(ascending=False)[:20].plot(kind="bar")

plt.yticks(range(0,190,10))
plt.show()
```
